```python
# Импортируем функцию разбиения данных на обучение и тест
from sklearn.model_selection import train_test_split

# Разделим данные на признаки и таргет
X = df_train.drop(columns='Attrition')
y = df_train['Attrition']
X_test = df_test


# Разделим данные на обучене и валидацию
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=123, stratify = y)
```

```python
# Закодируем выбранные категориальные признаки при помощи TargetEncoder
# T2
target_enc_t2 = TargetEncoder(verbose=1,
                           return_df=True,
                           cols=categ).fit(df_train.drop(columns=['id','Attrition']), df_train['Attrition'])
df_train_coded = target_enc_t2.transform(df_train.drop(columns=['id','Attrition']))
# Закодируем категориальные значения в тестовой выборке при помощи обученного на тренировочной части TargetEncoder.
df_test_coded = target_enc_t2.transform(df_test.drop(columns=['id']))

#T1
target_enc = TargetEncoder(verbose=1,
                            return_df=True,
                            cols=categ).fit(X_train.drop(columns=['id']), y_train)
X_train_coded = target_enc.transform(X_train.drop(columns=['id']))
# Закодируем категориальные значения в валидационнуй выборке при помощи обученного на тренировочной части TargetEncoder.
X_valid_coded = target_enc.transform(X_valid.drop(columns=['id']))
```

```python
# Подбор параметров
%%time

# Воспользуем функцией RandomizedSearchCV
# В качестве определения "лучшей" модели выберем f1_score

grid_LGBM = RandomizedSearchCV(
            lgb.LGBMClassifier(verbose=0),          # Алгоритм, в котором будем подбирать параметры
            param_distributions = {                 # Сетка параметров в виде словаря
                'max_depth': range(2, 7),
                'lr': np.linspace(0.001, 0.3, 100),
            },
            scoring = 'f1',
            cv = 3,                                 # CV для кросс-валидации (число или индексы)
            n_jobs = -1,                            # Число используемых ядер для работы
            return_train_score = True,              # Считать ли метрики на обучающей части
            n_iter=10,                              # Число итераций (сколько будем брать случайных наборов)
            verbose = 5,                            # Чем больше, тем алгоритм разговорчивее
        )

# Обучим сетку на обучающем датасете при помощи метода fit
grid_LGBM.fit(X_train_coded_scaled, y_train)
```

```python
# Выведем набор лучших параметров (они хранятся в best_params_) и лучшую метрику

print(f"Лучшая метрика:   {grid_LGBM.best_score_}")
print(f"Лучшие параметры: {grid_LGBM.best_params_}")
```

```python
# Получить отчет классификации передав в classification_report реальные метки и предсказанные
print("VALID")
print(classification_report(y_valid, grid_LGBM.best_estimator_.predict(X_valid_coded_scaled)))
print("TRAIN")
print(classification_report(y_train, grid_LGBM.best_estimator_.predict(X_train_coded_scaled)))

# Посмотреть на матрицу ошибок, передав в plot_confusion_matrix модель, тестовые данные и тестовые метки
plot_confusion_matrix(grid_LGBM.best_estimator_, X_valid_coded_scaled, y_valid)
plt.show()
```
